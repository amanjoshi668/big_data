```python
def get_data(party, state, local=False):
    search_bar = driver.find_element_by_css_selector(
            "input[placeholder='Search']")
    search_bar.clear()
    if local:
        save_state = state
        state = ""
    search_bar.send_keys(party + " " + state)
    search_bar.send_keys(Keys.RETURN)
    driver.implicitly_wait(30)
    python_button = driver.find_element_by_class_name("_6xu6")
    python_button.click(
    )  # x = driver.find_element_by_id("u_h_2")#.getText()
    driver.implicitly_wait(30)
    timestamps = driver.find_element_by_class_name("_5ptz")
    time = timestamps.get_attribute('title')
    video_url = driver.find_element_by_class_name('_400z')
    video_url = video_url.get_attribute('href')
    times.append(time)
    urls.append(video_url)
    if state == "":
        state = save_state
    df.loc[len(df)] = [state, party, driver.title,driver.current_url, time, video_url]
```

```python
times = []
urls = []
df = pd.DataFrame(columns=['state', 'party','fb_title', 'fb_url', 'time', 'video_url'])
counter = 0
for state in [
        "Telangana",
        "Uttarakhand",
        'West Bengal',
        'Odisha',
        'Sikkim',
        'Uttar Pradesh',
        'Tripura',
        'Maharastra',
        'Nagaland',
        'Mizoram',
        'Manipur',
        'Meghalya',
        'Nagaland',
        'Jammu and Kashmir',
        'Chattisgarh',
        'Andhra Pradesh',
        'Arunachal Pradesh'
]:
    for party in ["INC", "BJP"]:
        get_data(party, state)
```
